**Заданная функция**

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym

x = sym.Symbol('x')
y = sym.Symbol('y')
x = np.arange(-150, 151)
y = -12*x**4*np.sin(np.cos(x)) - 18*x**3+5*x**2 + 10*x - 30
```

**Построить график**

```python
fig, ax = plt.subplots()

ax.plot(x, y)
ax.grid()
plt.show()
```

**Определить корни (некоторые корни, но их бесконечное количество):**

```python
sym.pprint(sym.FiniteSet(x, y), use_unicode=True)
```

**Вывод:**
* 1_Корней бесконечное количество, на интервале, который неопределен;
* 2,3_Интервалов возрастания/убывания (между точками перегиба, где график первой производной функции пересекает ось x) также бесконечное количество;
* 4_График выше, на учатске x от -150 до 150;
* 5_Вершин также также бесконечное множество;
* 6,7_Промежутков на которой функция больше/меньше 0 (определяется знаком второй проиводной функции в точках перегиба) также бесконечное количество.
